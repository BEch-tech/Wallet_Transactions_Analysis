```python
import pandas as pd
```

```python
clean_df = pd.read_csv('C:/Users/beche/Documents/Grad Data Project/digital_wallet_transactions.csv')
clean_df
```

```python
clean_df.head()
```

```python
clean_df.columns
```

```python
columns_to_drop = ['transaction_id', 'user_id', 'merchant_id', 'idx']
columns_to_drop
```

```python
clean_df = clean_df.drop(columns=columns_to_drop)
clean_df.head()
```

```python
clean_df['transaction_date'] = pd.to_datetime(clean_df['transaction_date'])
```

```python
clean_df.info()
```

```python
clean_df.isnull()
```

```python
clean_df.isna()
```

```python
spending_patterns = clean_df.groupby('product_category') ['product_amount'].sum().reset_index()
spending_patterns
```

```python
cashback_effectiveness = clean_df.groupby('product_category') ['cashback'].mean().reset_index()
cashback_effectiveness
```

```python
loyalty_effectiveness = clean_df.groupby('product_category') ['loyalty_points'].mean().reset_index()
loyalty_effectiveness
```

```python
merchant_use = clean_df.groupby('merchant_name') ['product_amount'].sum().reset_index()
merchant_use
```

```python
merchant_use.rename(columns={'product_amount': 'merchant_use_total'}, inplace=True)
loyalty_effectiveness.rename(columns={'product_amount': 'loyalty_effectiveness_avg'}, inplace=True)
cashback_effectiveness.rename(columns={'product_amount': 'cashback_effectiveness_avg'}, inplace=True)
spending_patterns.rename(columns={'product_amount': 'max_spending_patterns'}, inplace=True)
```

```python
merchant_use.head()
```

```python
with pd.ExcelWriter('cleaned_digital_data.xlsx') as writer:
    clean_df.to_excel(writer, sheet_name='Original Data', index=False)
    merchant_use.to_excel(writer, sheet_name='Merchant Usage', index=False)
    loyalty_effectiveness.to_excel(writer, sheet_name='Loyalty Effectiveness', index=False)
    cashback_effectiveness.to_excel(writer, sheet_name='Cashback Effectiveness', index=False)
    spending_patterns.to_excel(writer, sheet_name='Spending Patterns', index=False)
```
